```python
import pandas as pd
import numpy as np

# Load your dataset
file_path =  r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_merged_stratified.csv"
data = pd.read_csv(file_path)  # Replace with the path to your dataset

# Parameters (adjustable)
alpha = 1  # Adjust according to importance of within-group inequality

# Grouping by location, sex, and age strata
grouped = data.groupby(['location_name', 'sex_name', 'age_name'])

# Function to calculate within-group inequality and retain SDI_quintile
def calculate_within_group_inequality(group):
    ng = len(group)  # Number of individuals in the group
    mu_g = group['DALY'].mean()  # Mean DALY for the group
    weights = group['SDI']  # Weight (SDI or other variable)
    
    # Calculating the term inside the summation for each individual
    weighted_term = (weights * (group['DALY'] / mu_g) * np.log(group['DALY'] / mu_g)).sum()
    
    # Normalizing by ln(ng)
    inequality_term = weighted_term / np.log(ng) if ng > 1 else 0  # Avoid division by zero
    
    # Select the most frequent SDI_quintile (or any other logic as needed)
    sdi_quintile = group['SDI_quintile'].mode().values[0] if not group['SDI_quintile'].isna().all() else np.nan

    # Returning the final inequality score and SDI_quintile
    return pd.Series({'Within_Group_Inequality': (alpha / 2) * inequality_term, 'SDI_quintile': sdi_quintile})

# Apply the function to calculate within-group inequality for each location-sex-age group
within_group_inequality = grouped.apply(calculate_within_group_inequality).reset_index()

# Save the results to a new CSV file
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within_group_inequality_dengue.csv"
within_group_inequality.to_csv(output_path, index=False)

# Display the results
print(within_group_inequality.head())
```

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_merged_stratified.csv"
df = pd.read_csv(file_path)

# Define the alpha parameter
ALPHA = 0.5  # Adjust as needed

# Ensure required columns exist
required_columns = ['sex_name', 'age_name', 'SDI_quintile', 'year', 'DALY', 'SDI']
missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    raise ValueError(f"Missing required columns in the dataset: {missing_columns}")

# Group data by (sex, age, SDI_quintile, year)
grouped = df.groupby(['sex_name', 'age_name', 'SDI_quintile', 'year'])

# Function to calculate within-group inequality
def calculate_within_group_inequality(group):
    ng = len(group)  # Number of individuals in the group
    if ng == 1:  # If only one data point, inequality is undefined
        return pd.Series({'Within_Group_Inequality': np.nan})

    mu_g = group['DALY'].mean()  # Mean DALY for the group
    weights = group['SDI']  # Assuming SDI is the weight (w_{gi})

    # Compute the weighted inequality term
    weighted_term = (weights * (group['DALY'] / mu_g) * np.log(group['DALY'] / mu_g)).sum()

    # Normalize by ln(n_g)
    inequality_term = (ALPHA / 2) * (weighted_term / np.log(ng)) if ng > 1 else 0  

    return pd.Series({'Within_Group_Inequality': inequality_term})

# Apply function to each group
df_with_inequality = df.merge(
    grouped.apply(calculate_within_group_inequality).reset_index(),
    on=['sex_name', 'age_name', 'SDI_quintile', 'year']
)

# Save the new dataset with within-group inequality values
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within1_group_inequality.csv"
df_with_inequality.to_csv(output_path, index=False)

print(f"Processed dataset saved to: {output_path}")
```

```python
import os
import pandas as pd
import numpy as np

# Load the dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_merged_stratified.csv"
data = pd.read_csv(file_path)

# Check if required columns exist
required_columns = ['location_name', 'age_name', 'sex_name', 'DALY', 'SDI', 'SDI_quintile']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Step 1: Calculate Within-Group Inequality Variance
def calculate_within_group_variance(group):
    """Calculates variance of within-group inequality for a given sex-age-country group."""
    mean_daly = group['DALY'].mean()
    within_variance = ((group['DALY'] / mean_daly) * np.log(group['DALY'] / mean_daly)).var()
    return within_variance

within_group_variance = (
    data.groupby(['location_name', 'age_name', 'sex_name'])
    .apply(calculate_within_group_variance)
    .mean()
)

# Step 2: Function to calculate between-group inequality for each country-age-sex group
def calculate_between_group_inequality_by_age_sex(group):
    """Calculates between-group inequality for a given country-age-sex group."""
    overall_mean = group['DALY'].mean()
    strata = group.groupby('SDI_quintile')

    strata_means = strata['DALY'].mean()  # μ_k
    strata_weights = strata['SDI'].mean()  # w_k (average SDI as weight)

    numerator = (strata_weights * ((strata_means - overall_mean) ** 2)).sum()
    denominator = overall_mean ** 2

    between_group_inequality = numerator / denominator  # Remove (beta/2) for now
    return pd.Series({'Between_Group_Inequality': between_group_inequality, 'SDI_quintile': group['SDI_quintile'].iloc[0]})

# Compute between-group inequality by country-age-sex while keeping SDI_quintile
between_group_inequality_by_age_sex = (
    data.groupby(['location_name', 'age_name', 'sex_name'])
        .apply(calculate_between_group_inequality_by_age_sex)
        .reset_index()
)

# Step 3: Calculate Between-Group Inequality Variance
between_group_variance = between_group_inequality_by_age_sex['Between_Group_Inequality'].var()

# Step 4: Compute Alpha and Beta Dynamically
alpha = within_group_variance / (within_group_variance + between_group_variance)
beta = between_group_variance / (within_group_variance + between_group_variance)

print(f"Calculated Alpha: {alpha:.4f}, Beta: {beta:.4f}")

# Step 5: Apply beta to the between-group inequality calculation
between_group_inequality_by_age_sex['Between_Group_Inequality'] *= beta / 2

# Ensure the output directory exists
output_dir =r"C:\Users\Shekoofeh\Downloads\NTD\Dengue"
os.makedirs(output_dir, exist_ok=True)

# Save results
output_path = os.path.join(output_dir, "between_group_inequality_by_country_age_sex.csv")
between_group_inequality_by_age_sex.to_csv(output_path, index=False)

# Display results
print(between_group_inequality_by_age_sex.head())
```

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within1_group_inequality.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
required_columns = ['location_name', 'age_name', 'sex_name', 'DALY', 'SDI', 'SDI_quintile']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Define beta (adjustable weight for between-group importance)
beta = 1  # Adjust as needed

# Step 1: Calculate Overall Mean DALY per group
overall_mean_daly = data.groupby(['location_name', 'age_name', 'sex_name'])['DALY'].mean().reset_index()
overall_mean_daly.rename(columns={'DALY': 'Overall_Mean_DALY'}, inplace=True)

# Merge back to ensure all rows retain their overall mean
data = data.merge(overall_mean_daly, on=['location_name', 'age_name', 'sex_name'], how='left')

# Step 2: Compute mean DALY and weights per SDI quintile
strata_mean_daly = data.groupby(['location_name', 'age_name', 'sex_name', 'SDI_quintile'])['DALY'].mean().reset_index()
strata_mean_daly.rename(columns={'DALY': 'Strata_Mean_DALY'}, inplace=True)

strata_weights = data.groupby(['location_name', 'age_name', 'sex_name', 'SDI_quintile'])['SDI'].mean().reset_index()
strata_weights.rename(columns={'SDI': 'Strata_Weight_SDI'}, inplace=True)

# Merge back to the main dataset
data = data.merge(strata_mean_daly, on=['location_name', 'age_name', 'sex_name', 'SDI_quintile'], how='left')
data = data.merge(strata_weights, on=['location_name', 'age_name', 'sex_name', 'SDI_quintile'], how='left')

# Step 3: Calculate Between-Group Inequality
def calculate_between_group_inequality(group):
    overall_mean = group['Overall_Mean_DALY'].iloc[0]
    n_g = len(group)  # Number of observations in the group

    if overall_mean == 0 or n_g == 1:  # Avoid division by zero
        return 0

    # Compute SDI-weighted variance
    strata_means = group.groupby('SDI_quintile')['Strata_Mean_DALY'].mean()
    strata_weights = group.groupby('SDI_quintile')['Strata_Weight_SDI'].mean()

    numerator = (strata_weights * (strata_means - overall_mean) ** 2).sum()
    denominator = overall_mean ** 2

    # Apply formula with β factor and normalization
    between_group_inequality = (beta / 2) * (numerator / denominator) * np.log(n_g)

    return between_group_inequality if denominator != 0 else 0  # Ensure no division by zero

# Compute between-group inequality for each location-age-sex group
between_group_results = data.groupby(['location_name', 'age_name', 'sex_name']).apply(
    calculate_between_group_inequality
).reset_index()

# Rename and merge results back
between_group_results.rename(columns={0: 'Between_Group_Inequality'}, inplace=True)
data = data.merge(between_group_results, on=['location_name', 'age_name', 'sex_name'], how='left')

# Fill any missing values with 0
data['Between_Group_Inequality'].fillna(0, inplace=True)

# Save results
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within_between_group_inequality.csv"
data.to_csv(output_path, index=False)

print(f"Processed dataset saved to: {output_path}")
```

```python
import pandas as pd

# File paths for input files
within_group_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within_group_inequality_dengue.csv"
between_group_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\between_group_inequality_by_country_age_sex.csv"

# Load the within-group and between-group inequality data
within_group_df = pd.read_csv(within_group_path)
between_group_df = pd.read_csv(between_group_path)

# Merge the dataframes on location_name, age_name, sex_name, and SDI_quintile
merged_df = pd.merge(
    within_group_df,
    between_group_df,
    on=['location_name', 'age_name', 'sex_name', 'SDI_quintile'],
    how='inner'
)

# Calculate the variance of Within-Group and Between-Group inequalities
var_within = merged_df['Within_Group_Inequality'].var()
var_between = merged_df['Between_Group_Inequality'].var()

# Dynamically calculate alpha and beta
alpha = var_within / (var_within + var_between)
beta = var_between / (var_within + var_between)

# Calculate CISI for each row
merged_df['CISI'] = (alpha / 2) * merged_df['Within_Group_Inequality'] + (beta / 2) * merged_df['Between_Group_Inequality']

# Aggregate results by SDI_quintile (SDI Tier)
summary_df = merged_df.groupby('SDI_quintile').agg(
    Mean_Within_Group_Inequality=('Within_Group_Inequality', 'mean'),
    Mean_Between_Group_Inequality=('Between_Group_Inequality', 'mean'),
    Mean_CISI=('CISI', 'mean')
).reset_index()

# Rename SDI_quintile to SDI Tier
summary_df.rename(columns={'SDI_quintile': 'SDI Tier'}, inplace=True)

# Save the final aggregated results to a new CSV file
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\CISI_Dengue_Summary.csv"
summary_df.to_csv(output_path, index=False)

# Display results
print(summary_df)
```

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_within_between_group_inequality.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
required_columns = ['location_name', 'age_name', 'sex_name', 'DALY', 'SDI', 'SDI_quintile', 
                    'Within_Group_Inequality', 'Between_Group_Inequality']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Define alpha and beta (adjustable weights)
alpha = 1  # Adjust as needed
beta = 1  # Already used in previous computations

# Step 1: Compute first term (Within-Group Component)
def compute_within_group_component(group):
    n_g = len(group)  # Number of individuals in the group
    if n_g == 1:
        return 0  # Avoid division by zero
    
    # Mean DALY per group
    group_mean = group['DALY'].mean()
    
    # Compute the summation term in the formula
    weighted_entropy = ((group['DALY'] / group_mean) * np.log(group['DALY'] / group_mean) * group['SDI']).sum()
    
    # Normalize by ln(n_g)
    within_component = (alpha / 2) * (weighted_entropy / np.log(n_g))
    
    return within_component if np.isfinite(within_component) else 0  # Handle NaN or infinite values

# Compute Within-Group CISI component per (location, age, sex)
within_group_results = data.groupby(['location_name', 'age_name', 'sex_name']).apply(
    compute_within_group_component
).reset_index()

# Rename and merge results back
within_group_results.rename(columns={0: 'Within_Group_CISI_Component'}, inplace=True)
data = data.merge(within_group_results, on=['location_name', 'age_name', 'sex_name'], how='left')

# Step 2: Compute CISI for each row
data['CISI'] = data['Within_Group_CISI_Component'] + data['Between_Group_Inequality']

# Save results
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\Dengue_CISI_2.csv"
data.to_csv(output_path, index=False)

print(f"CISI results saved to: {output_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
file_path =  r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\Dengue_CISI_2.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
required_columns = ['Within_Group_Inequality', 'Between_Group_Inequality', 'CISI']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Number of Monte Carlo simulations
num_simulations = 500  

# Define percentage variation (±10%)
variation_percentage = 0.1  

# Generate random noise matrix for all three features
perturbation_factors = np.random.normal(1, variation_percentage, (num_simulations, len(data)))

# Apply perturbations
perturbed_within = data['Within_Group_Inequality'].values * perturbation_factors
perturbed_between = data['Between_Group_Inequality'].values * perturbation_factors
perturbed_CISI = perturbed_within + perturbed_between  # Recomputed CISI

# Calculate summary statistics
mean_within = perturbed_within.mean(axis=0)
std_within = perturbed_within.std(axis=0)

mean_between = perturbed_between.mean(axis=0)
std_between = perturbed_between.std(axis=0)

mean_CISI = perturbed_CISI.mean(axis=0)
std_CISI = perturbed_CISI.std(axis=0)

# Save summary results
summary_df = pd.DataFrame({
    'Mean_Within_Group_Inequality': mean_within,
    'Std_Within_Group_Inequality': std_within,
    'Mean_Between_Group_Inequality': mean_between,
    'Std_Between_Group_Inequality': std_between,
    'Mean_CISI': mean_CISI,
    'Std_CISI': std_CISI
})
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\monte_carlo_CISI_summary.csv"
summary_df.to_csv(output_path, index=False)

print(f"Monte Carlo summary saved to: {output_path}")

# Plot histograms for all three features
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Within-Group Inequality Distribution
axes[0].hist(mean_within, bins=40, color='blue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Mean Within-Group Inequality')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Monte Carlo - Within-Group Inequality')
axes[0].grid(True)

# Between-Group Inequality Distribution
axes[1].hist(mean_between, bins=40, color='red', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Mean Between-Group Inequality')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Monte Carlo - Between-Group Inequality')
axes[1].grid(True)

# CISI Distribution
axes[2].hist(mean_CISI, bins=40, color='green', edgecolor='black', alpha=0.7)
axes[2].set_xlabel('Mean CISI')
axes[2].set_ylabel('Frequency')
axes[2].set_title('Monte Carlo - CISI')
axes[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\Dengue_CISI_2.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
required_columns = ['Within_Group_Inequality', 'Between_Group_Inequality', 'CISI']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Number of Monte Carlo simulations
num_simulations = 500  

# Define percentage variation (±10%)
variation_percentage = 0.1  

# Generate random noise matrix for all three features
perturbation_factors = np.random.normal(1, variation_percentage, (num_simulations, len(data)))

# Apply perturbations
perturbed_within = data['Within_Group_Inequality'].values * perturbation_factors
perturbed_between = data['Between_Group_Inequality'].values * perturbation_factors
perturbed_CISI = perturbed_within + perturbed_between  # Recomputed CISI

# Calculate summary statistics
mean_within = perturbed_within.mean(axis=0)
std_within = perturbed_within.std(axis=0)

mean_between = perturbed_between.mean(axis=0)
std_between = perturbed_between.std(axis=0)

mean_CISI = perturbed_CISI.mean(axis=0)
std_CISI = perturbed_CISI.std(axis=0)

# Save summary results
summary_df = pd.DataFrame({
    'Mean_Within_Group_Inequality': mean_within,
    'Std_Within_Group_Inequality': std_within,
    'Mean_Between_Group_Inequality': mean_between,
    'Std_Between_Group_Inequality': std_between,
    'Mean_CISI': mean_CISI,
    'Std_CISI': std_CISI
})
output_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\monte_carlo_CISI_summary.csv"
summary_df.to_csv(output_path, index=False)

print(f"Monte Carlo summary saved to: {output_path}")

# Plot histograms for all three features
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Within-Group Inequality Distribution
axes[0].hist(mean_within, bins=40, color='blue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Mean Within-Group Inequality')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Monte Carlo - Within-Group Inequality')
axes[0].grid(True)

# Between-Group Inequality Distribution
axes[1].hist(mean_between, bins=40, color='red', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Mean Between-Group Inequality')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Monte Carlo - Between-Group Inequality')
axes[1].grid(True)

# CISI Distribution
axes[2].hist(mean_CISI, bins=40, color='green', edgecolor='black', alpha=0.7)
axes[2].set_xlabel('Mean CISI')
axes[2].set_ylabel('Frequency')
axes[2].set_title('Monte Carlo - CISI')
axes[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load dataset
file_path = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\dengue_merged_stratified.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
required_columns = ['DALY', 'cause_name', 'SDI_quintile', 'age_name', 'sex_name']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Descriptive statistics for DALYs grouped by disease, SDI quintile, age, and sex
descriptive_stats = data.groupby(['cause_name', 'SDI_quintile', 'age_name', 'sex_name'])['DALY'].describe()

# Calculate mean DALYs per disease, SDI tier, age, and sex
mean_dalys = data.groupby(['cause_name', 'SDI_quintile', 'age_name', 'sex_name'])['DALY'].mean().reset_index()

# Pivot table for better readability
mean_dalys_pivot = mean_dalys.pivot_table(
    index=['cause_name', 'age_name', 'sex_name'],
    columns='SDI_quintile',
    values='DALY'
)

# Save descriptive statistics and mean DALY table to CSV
descriptive_stats_output = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\descriptive_stats_extended.csv"
mean_dalys_output = r"C:\Users\Shekoofeh\Downloads\NTD\Dengue\mean_dalys_extended.csv"
descriptive_stats.to_csv(descriptive_stats_output)
mean_dalys_pivot.to_csv(mean_dalys_output)

print(f"Descriptive statistics saved to: {descriptive_stats_output}")
print(f"Mean DALYs per SDI, age, and sex saved to: {mean_dalys_output}")

# Print summary table
print("\nMean DALYs by Disease, Age, and Sex Across SDI Tiers:")
print(mean_dalys_pivot)
```
